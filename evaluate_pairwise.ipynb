```python
# a = false positive
# b = false negative

def error_analysis(magazine, margin):
    output = []
    for k in magazine:
        page = {'page': k['p'], 'ca': 0, 'cb': 0, 'ia': 0, 'ib': 0}
        for j in k['data']:
            if j['label'] == 0:
                if j['value'] >= margin:
                    page['ib'] += 1
                else:
                    page['cb'] += 1
            else:
                if j['value'] >= margin:
                    page['ca'] += 1
                else:
                    page['ia'] += 1
        output.append(page)
    return output
```

```python
def calc_total_error(goldeneblatt, bilderfrau, margin): 
    errors = error_analysis(goldeneblatt, margin)
    t = 0
    c = 0
    for err in errors: 
        t += err['ca'] + err['cb'] + err['ia'] + err['ib']
        c += err['ca'] + err['cb']
        
    errors2 = error_analysis(bilderfrau, margin)
    t2 = 0
    c2 = 0
    for err in errors2: 
        t2 += err['ca'] + err['cb'] + err['ia'] + err['ib']
        c2 += err['ca'] + err['cb']
        
    return {'GB': {
        'c': c,
        't':t
    },
            'BF': {
                'c': c2,
                't': t2
            }}
    

```

```python
def calc_perc(o): 
    return [o['GB']['c']/o['GB']['t'],o['BF']['c']/o['BF']['t']]
```
